```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

* Mean Absolute Error(MAE)
* Mean Squared Error (MSE)
* R2

```python
x = np.random.normal (10, 2 , 120)
y = x + np.random.normal(3, 0.5 ,120)
```

```python
x[:5]
```

```python
y[:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size = 0.2, 
                                                    shuffle = True, 
                                                    random_state = 666)
```

```python
plt.scatter (x_train, y_train, c='b', label = 'Training Data')
plt.scatter (x_test, y_test, c= 'r', label = 'Testing Data')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
m,c,r2 , _, err = linregress(x_train, y_train)
```

```python
m
```

```python
c
```

```python
r2
```

```python
reg_line_train = m* x_train + c 
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, reg_line_train, c='r')
plt.show()
```

```python
reg_line_test = m* x_test + c
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, reg_line_test, c='r')
plt.show()
```

```python
def predict(x_new):
    return m* x_new +c
```
